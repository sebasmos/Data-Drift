```python
def bootstrap_auroc_diff(df, ethnicity, nboot=4000):
    df_eth = df[df['ethnicity'] == ethnicity]
    
    df_2014 = df_eth[df_eth['set'] == '2014-2015']
    df_2020 = df_eth[df_eth['set'] == '2020-2021']
    
    y_true_2014 = df_2014['mortality'].values
    y_pred_2014 = df_2014['predicted_hospital_mortality'].values
    
    y_true_2020 = df_2020['mortality'].values
    y_pred_2020 = df_2020['predicted_hospital_mortality'].values
    
    n_2014 = len(y_true_2014)
    n_2020 = len(y_true_2020)
    
    diffs = []
    
    for _ in range(nboot):
        idx_2014 = np.random.randint(0, n_2014, n_2014)
        idx_2020 = np.random.randint(0, n_2020, n_2020)
        
        y_true_2014_bs = y_true_2014[idx_2014]
        y_pred_2014_bs = y_pred_2014[idx_2014]
        
        y_true_2020_bs = y_true_2020[idx_2020]
        y_pred_2020_bs = y_pred_2020[idx_2020]
        
        if len(np.unique(y_true_2014_bs)) < 2 or len(np.unique(y_true_2020_bs)) < 2:
            continue 
        
        auroc_2014 = roc_auc_score(y_true_2014_bs, y_pred_2014_bs)
        auroc_2020 = roc_auc_score(y_true_2020_bs, y_pred_2020_bs)
        
        diffs.append(auroc_2020 - auroc_2014)
    
    return np.array(diffs)

results = []
for race in df_sepsis['ethnicity'].unique():
    diff_bootstrap = bootstrap_auroc_diff(df_sepsis, race, nboot=4000)
    mean_diff = np.mean(diff_bootstrap)
    ci_lower = np.percentile(diff_bootstrap, 2.5)
    ci_upper = np.percentile(diff_bootstrap, 97.5)
    
    results.append({
        'ethnicity': race,
        'mean_diff_auroc': mean_diff,
        'ci_lower': ci_lower,
        'ci_upper': ci_upper,
        'bootstrap_distribution': diff_bootstrap  
    })

auroc_diff_df = pd.DataFrame(results)
```

```python
def compute_smr(y_true, y_pred):
    observed = np.sum(y_true)
    expected = np.sum(y_pred)
    return observed / expected if expected > 0 else np.nan

def bootstrap_smr_distance_diff(df, ethnicity, nboot=4000):
    df_eth = df[df['ethnicity'] == ethnicity]
    df_2014 = df_eth[df_eth['set'] == '2014-2015']
    df_2020 = df_eth[df_eth['set'] == '2020-2021']
    
    y_true_2014 = df_2014['mortality'].values
    y_pred_2014 = df_2014['predicted_hospital_mortality'].values
    n_2014 = len(y_true_2014)
    
    y_true_2020 = df_2020['mortality'].values
    y_pred_2020 = df_2020['predicted_hospital_mortality'].values
    n_2020 = len(y_true_2020)
    
    diffs = []
    
    for _ in range(nboot):
        idx_2014 = np.random.randint(0, n_2014, n_2014)
        idx_2020 = np.random.randint(0, n_2020, n_2020)
        
        y_true_2014_bs = y_true_2014[idx_2014]
        y_pred_2014_bs = y_pred_2014[idx_2014]
        
        y_true_2020_bs = y_true_2020[idx_2020]
        y_pred_2020_bs = y_pred_2020[idx_2020]
        
        smr_2014 = compute_smr(y_true_2014_bs, y_pred_2014_bs)
        smr_2020 = compute_smr(y_true_2020_bs, y_pred_2020_bs)
        
        if np.isnan(smr_2014) or np.isnan(smr_2020):
            continue
        
        dist_2014 = abs(smr_2014 - 1)
        dist_2020 = abs(smr_2020 - 1)
        
        diffs.append(dist_2020 - dist_2014)
    
    return np.array(diffs)

results = []
for race in df_sepsis['ethnicity'].unique():
    diff_bootstrap = bootstrap_smr_distance_diff(df_sepsis, race, nboot=4000)
    mean_diff = np.mean(diff_bootstrap)
    ci_lower = np.percentile(diff_bootstrap, 2.5)
    ci_upper = np.percentile(diff_bootstrap, 97.5)
    
    results.append({
        'ethnicity': race,
        'mean_diff_smr_distance': mean_diff,
        'ci_lower': ci_lower,
        'ci_upper': ci_upper,
        'bootstrap_distribution': diff_bootstrap      })

smr_diff_df = pd.DataFrame(results)
```

```python
results = []

for race in df_sepsis['ethnicity'].dropna().unique():
    row = {'ethnicity': race}
    
    for set_val in df_sepsis['set'].unique():
        subset = df_sepsis[(df_sepsis['ethnicity'] == race) & (df_sepsis['set'] == set_val)]
        
        if len(subset) > 0:
            observed = subset['mortality'].sum()
            expected = subset['predicted_hospital_mortality'].sum()
            smr = observed / expected if expected > 0 else None
            row[set_val] = smr
        else:
            row[set_val] = None
            
    results.append(row)

smr_df = pd.DataFrame(results)
smr_df
```

```python
def bootstrap (x, nboot, statfun):
    values = np.array(x)
    resampled_stat = []
    
    for k in range(nboot):
        index = np.random.randint(0, len(x), len(x))
        sample = x[index]
        bstatistic = statfun(sample)
        resampled_stat.append(bstatistic)
        
    return np.array(resampled_stat)
```

```python
auroc_bootstrap = bootstrap(auroc_df['difference'], 4000, np.mean)
smr_bootstrap = bootstrap(smr_df['difference'], 4000, np.mean)
```

```python

def prepare_long_df(result_df):
    records = []
    for _, row in result_df.iterrows():
        for val in row['bootstrap_distribution']:
            records.append({
                'ethnicity': row['ethnicity'],
                'value': val
            })
    return pd.DataFrame(records)

def plot_violin_with_ci(long_df, ci_df, title, ylabel):
    sns.set(style="whitegrid")
    plt.figure(figsize=(12, 6))

    sns.violinplot(data=long_df, x='ethnicity', y='value', inner=None, color=".85")
    sns.stripplot(data=long_df, x='ethnicity', y='value', color='k', size=1, alpha=0.3, jitter=True)

    for i, row in ci_df.iterrows():
        plt.plot([i - 0.2, i + 0.2], [row['mean_diff'], row['mean_diff']], color='blue', linewidth=2)
        plt.plot([i, i], [row['ci_lower'], row['ci_upper']], color='black', linewidth=1.5)

    plt.axhline(0, color='red', linestyle='--', label='No Difference')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend()
    plt.show()

# --- Apply to AUROC ---
auroc_plot_df = auroc_diff_df.rename(columns={'mean_diff_auroc': 'mean_diff'})
auroc_long_df = prepare_long_df(auroc_plot_df)

plot_violin_with_ci(
    auroc_long_df,
    auroc_plot_df,
    title="Bootstrap AUROC Difference (20142015 vs 20202021)",
    ylabel="AUROC Difference"
)
```
